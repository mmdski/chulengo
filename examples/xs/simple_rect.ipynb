```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from chulengo.xs.rect import SimpleRect


width = 1
n = 0.04
left_ineff = 0.1
right_ineff = 0.9

xs = SimpleRect(width, n, left_ineff, right_ineff)
```

```python
h = np.linspace(-.1, 1)
_ = plt.plot(xs.area(h), h, label="Area")
_ = plt.plot(xs.flow_area(h), h, label="Flow Area")
_ = plt.legend()
_ = plt.xlabel("Area, in square feet")
_ = plt.ylabel("Stage, in feet")
```

```python
_ = plt.plot(xs.conveyance(h), h)
_ = plt.xlabel("Conveyance, in cfs")
_ = plt.ylabel("Stage, in feet")
```
